```python
import time
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
ppg_file = r'.\data\ppg.xlsx' 
```

```python
# 读取Excel文件
df = pd.read_excel(ppg_file)

vals = []
for col in df.columns:
    # print("col:", col)
    tmp = col.split(',')[1:-1]
    # print("tmp:", tmp)
    for val in tmp:
        vals.append(float(val))
print(vals)
```

```python
def plot_list(numbers):
    plt.plot(numbers)
    plt.show()
```

```python
plot_list(vals[0:500])
```

```python
ecg_file = r'.\data\ecg.tmp'
```

```python
# 打开并读取 tmp 文件
with open(ecg_file, 'r') as file:
    ecg_data = file.read()
    
# 输出内容
print(ecg_data)
```
