```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义文件夹路径
folder_path = "./.cache/1"
```

### 统一合并显示

```python
# 获取文件夹中的所有文件名
file_names = os.listdir(folder_path)

# 遍历文件名
for file_name in file_names:
    # 只处理以 .txt 结尾的文件
    if file_name.endswith(".txt"):
        # 构建文件路径
        file_path = os.path.join(folder_path, file_name)

        print(file_path)
        # 读取文件中的整数数字
        with open(file_path, "r") as file:
            numbers = [int(line.strip()) for line in file.readlines()]
        
        # 绘制折线图
        plt.plot(numbers, label=file_name[:-4])  # 去除文件名中的 .txt 后缀作为标签

# 添加图例
plt.legend()
# 显示图形
plt.show()
```

### 横坐标转化为时间进行显示

```python
# 计算时间间隔
sampling_rate = 100  # 采样频率为100Hz
time_interval = 1 / sampling_rate  # 时间间隔
```

```python
# 遍历文件名
for i, file_name in enumerate(file_names):
    # 只处理以 .txt 结尾的文件
    if file_name.endswith(".txt"):
        # 构建文件路径
        file_path = os.path.join(folder_path, file_name)
        
        # 读取文件中的整数数字
        with open(file_path, "r") as file:
            numbers = [int(line.strip()) for line in file.readlines()]
        
        # 计算时间
        times = np.arange(0, len(numbers) * time_interval, time_interval)
        
        # 创建子图
        plt.subplot(len(file_names), 1, i+1)
        
        # 绘制折线图
        plt.plot(times, numbers)
        
        # 设置标题
        plt.title(file_name[:-4])
        
        # 设置横纵坐标标签
        plt.xlabel("Time (s)")
        plt.ylabel("Value")

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

#### 部分显示

```python
# 计算时间点
time_points = np.arange(0, 20, time_interval)

# 遍历文件名
for i, file_name in enumerate(file_names):
    # 只处理以 .txt 结尾的文件
    if file_name.endswith(".txt"):
        # 构建文件路径
        file_path = os.path.join(folder_path, file_name)
        
        # 读取文件中的整数数字
        with open(file_path, "r") as file:
            numbers = [int(line.strip()) for line in file.readlines()]
        
        # 创建子图
        plt.subplot(len(file_names), 1, i+1)
        
        # 绘制折线图，仅绘制前20s的数据
        plt.plot(time_points, numbers[:len(time_points)])
        
        # 设置标题
        plt.title(file_name[:-4])
        
        # 设置横纵坐标标签
        plt.xlabel("Time (s)")
        plt.ylabel("Value")

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
